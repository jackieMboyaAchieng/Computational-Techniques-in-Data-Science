```python
from fpdf import FPDF

# Custom PDF class
class ThreatMapPDF(FPDF):
    def header(self):
        self.set_font("Arial", "B", 14)
        self.cell(0, 10, "Mars University Network Threat Map", ln=True, align="C")
        self.ln(5)

    def add_threat(self, title, color, description):
        self.set_fill_color(*color)
        self.set_text_color(0)
        self.set_font("Arial", "B", 12)
        self.cell(0, 10, title, ln=True, fill=True)
        self.set_font("Arial", "", 11)
        self.multi_cell(0, 8, description)
        self.ln(2)

# Threat data
threats = [
    {
        "title": "Malware Infection via BYOD",
        "color": (255, 200, 200),
        "desc": "Personal devices from students and staff can introduce malware to internal systems, particularly in the School of Computing and Human Resources."
    },
    {
        "title": "DDoS Vulnerability",
        "color": (255, 255, 200),
        "desc": "The exposed connection to the cloud/internet via Router0 is a critical point susceptible to DDoS attacks targeting availability."
    },
    {
        "title": "Weak Access Controls in Server Room",
        "color": (200, 220, 255),
        "desc": "Lack of network segmentation and outdated authentication practices leave FTP, Mail, and Web servers vulnerable to unauthorized access."
    },
    {
        "title": "Insider Threats",
        "color": (200, 255, 200),
        "desc": "Employees in Administration and Human Resources with elevated privileges pose insider risks due to poor monitoring and audit mechanisms."
    },
    {
        "title": "Cloud Misconfiguration",
        "color": (230, 200, 255),
        "desc": "Improperly configured cloud access points can result in data leakage, especially affecting external research collaborations."
    },
    {
        "title": "Lack of Segmentation",
        "color": (255, 230, 180),
        "desc": "Cross-department routing without strict firewall rules (e.g., between School of Computing and Server Room) increases lateral movement risk."
    }
]

# Create PDF
pdf = ThreatMapPDF()
pdf.add_page()
for threat in threats:
    pdf.add_threat(threat["title"], threat["color"], threat["desc"])

# Save file
pdf.output("Mars_University_Threat_Map.pdf")
```

```python
class RansomwareAttack:
    def __init__(self, target):
        self.target = target
        self.infected_devices = []
        self.access_level = 'user'
        self.network_access = False
        self.critical_servers_compromised = False
        self.files_encrypted = False

    def phishing_email(self, victim):
        print(f"Phishing email sent to {victim} with malicious attachment.")
        self.open_malicious_attachment(victim)

    def open_malicious_attachment(self, victim):
        print(f"{victim} opened malicious attachment. Device infected with ransomware.")
        self.infected_devices.append(victim)
        self.escalate_privileges()

    def escalate_privileges(self):
        print("Attacker escalates privileges from user to admin.")
        self.access_level = 'admin'
        self.lateral_movement()

    def lateral_movement(self):
        print("Attacker moves laterally across the network due to weak segmentation.")
        self.network_access = True
        self.access_critical_servers()

    def access_critical_servers(self):
        print("Attacker accesses critical research servers containing sensitive AI and space exploration data.")
        self.critical_servers_compromised = True
        self.encrypt_files()

    def encrypt_files(self):
        print("Ransomware encrypts critical files, making them inaccessible.")
        self.files_encrypted = True
        self.demand_ransom()

    def demand_ransom(self):
        print("Attacker demands payment in cryptocurrency to restore access.")
        self.impact_assessment()

    def impact_assessment(self):
        print("Severe impacts: Operational disruption, reputational damage, financial losses, and research delays.")


# Simulate the ransomware attack on Mars University
attack_simulation = RansomwareAttack(target="Mars University")
attack_simulation.phishing_email(victim="faculty_member@marsuniversity.edu")
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Define the steps of the ransomware attack
steps = [
    "Phishing Email Sent",
    "Faculty Opens Malicious Attachment",
    "Ransomware Installation",
    "Privilege Escalation",
    "Lateral Movement in Network",
    "Access Critical Research Servers",
    "Encrypt Sensitive Data",
    "Demand Cryptocurrency Ransom",
    "Impact: Financial, Reputation, Research Delays"
]

# Define relationships between steps (edges)
edges = [
    ("Phishing Email Sent", "Faculty Opens Malicious Attachment"),
    ("Faculty Opens Malicious Attachment", "Ransomware Installation"),
    ("Ransomware Installation", "Privilege Escalation"),
    ("Privilege Escalation", "Lateral Movement in Network"),
    ("Lateral Movement in Network", "Access Critical Research Servers"),
    ("Access Critical Research Servers", "Encrypt Sensitive Data"),
    ("Encrypt Sensitive Data", "Demand Cryptocurrency Ransom"),
    ("Demand Cryptocurrency Ransom", "Impact: Financial, Reputation, Research Delays")
]

# Create directed graph
G = nx.DiGraph()
G.add_edges_from(edges)

# Define layout for better visualization
pos = nx.spring_layout(G, seed=42)

# Plot the flowchart
plt.figure(figsize=(12, 8))
nx.draw(G, pos, with_labels=True, node_size=4500, node_color="skyblue", font_size=10, font_weight="bold", arrowsize=20, arrowstyle='-|>')

# Display flowchart
plt.title("Flowchart of Ransomware Attack on Mars University")
plt.show()
```
