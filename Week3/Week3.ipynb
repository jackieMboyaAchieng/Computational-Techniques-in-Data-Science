## Problem
A factory produces two types of products: Product A and Product B. Thep rofit for Product A is $20 per unit and for Product B is $30 per unit
(Manufacturing Problem)
- Product A requires 2 hours of labor and I unit of raw material
- Product B requires 1 hour of labor and 2 units of raw material.
- Available Labor = 100 hours.
- Available Raw Material = 80 units.

Maximize the profit.

```python
from scipy.optimize import linprog
```

```python
#Objective Functions
c=[-20,-30]
```

```python
# Inequalities (Constraints)
A = [[2,1],[1,2]]
b = [100,80]
```

```python
x_bounds = (0,None)
y_bounds = (0,None)
```

```python
#Solve the LP problem
Results = linprog(c,A_ub=A,b_ub=b,bounds=[x_bounds,y_bounds], method="highs")
Results
```

```python
#Output
print("Optimal number of Product A:", round(Results.x[0]))
```

```python
print("Optimal number of Product B:", round(Results.x[1]))
```

```python
print("Optimal Profit:", -round(Results.fun))
```

## Problem 2: Transport Optimization

A company ships goods from two warehouses (A, B) to two markets (X,
Y) (Transport Optimization).
- Cost from X $4, A to Y $5
- Cost from B to X $6, B to Y $3
- Warehouse A capacity 70 units
- Warehouse B capacity 50 units
- Market X demand 60 units
- Market Y demand 50 units.

Minimize the transportation cost.

Solution

Costs:

- A to X: $4 per unit
- A to Y: $5 per unit
- B to X: $6 per unit
- B to Y: $3 per unit

Capacities:

- Warehouse A: 70 units
- Warehouse B: 50 units

Demands:
- Market X: 60 units
- Market Y: 50 units

Objective:
- Minimize the total transportation cost.

```python
## Objective Function
c_min=[4,5,6,3]
```

```python
#Constraints
A=[[1,1,0,0], #Warehouse A capacity
   [0,0,1,1], #Warehouse B capacity
   [1,0,1,0], #Market X Demand
   [0,1,0,1]] #Marke Y Demand

b=[70,50,60,50]
```

```python
Result_Min=linprog(c_min,A_eq=A,b_eq=b, bounds=(0,None), method="highs")
Result_Min
```

```python
print("Minimun Transport Cost:", round(Result_Min.fun))
```

## Problem 3
A person needs at least 2000 calories and 50g protein daily (Diet
Optimization)
- Food A: 500 calories, 30g protein, `$3/` unit.
- Food B: 700 calories, 20g protein, `$5/` unit.

```python
# objective function
c_opt = [3, 5]
```

```python
# constraints
A = [[-500, -700], # Calories constraints
 [-30, -20]] # Protein constraints
B = [-2000, -50]
```

```python
# solve the problem
Results_Opt = linprog(c_opt, A_ub=A, b_ub=B, bounds=(0,None),method = 'highs')
```

```python
Results_Opt
```

```python
print("Optimal number of Chicken Breast:", round(Results_Opt.x[0]))
```

```python
print("Optimal number of Tofu:", round(Results_Opt.x[1]))
```

```python
print("Manimum calories:", round(Results_Opt.fun))
```

## Problem 4: Minimize the cost.
An investor has $100,000 to invest in two stocks
(Portfolio Optimization)
- Stock A: Expected return — 8%, risk 5%.
- Stock B: Expected return — 12%, risk 10%.

The investor wants to minimize risk while ensuring a 10% return.

```python
# Objective Function
c_risk = [5,10]
```

```python
#Constraints
A=[[-8,-12], #Return Constraint
   [1,1]] #Total Investment

b=[-10000,100000]
```

```python
Result_risks=linprog(c_risk,A_ub=A,b_ub=b,bounds=(0,None),method="highs")
Result_risks
```

```python
print("Minimum Risk:", round(Result_risks.fun))
```

```python
#Investment in stock A
print("Investments in Stock A:", round(Result_risks.x[0]))
```

```python
#Investment in stock B
print("Investments in Stock B:", round(Result_risks.x[1]))
```
