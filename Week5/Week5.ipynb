```python
import numpy as np
```

```python
# Transition matrix in a Markov process (rows = current state, columns = next state)
P = np.array([
    [0.5, 0.3, 0.2], # Rainy
    [0.2,0.6,0.2], # Cloudy
    [0.1, 0.3, 0.6] # Sunny
])
P
```

```python
# Chacking stationarity
def is_stationary(P):
  # return sum
  return np.allclose(P.sum(axis=1), 1)
print(is_stationary(P))
```
