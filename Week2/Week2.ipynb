```python
#import
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Random walk one-dimensional
num_steps=1000 #Number of steps
steps=np.random.choice([-1,1], size=num_steps) #Random choice -1 and 1
positions=np.cumsum(steps) #Cumulative sum of steps (to get position)
```

```python
#plot
plt.figure(figsize=(10,6))
plt.plot(positions, label="1D Random Walk")
plt.axhline(y=0, color='black', linestyle='--', linewidth=0.8)
plt.xlabel('Steps')
plt.ylabel('Position')
plt.title('1D Random Walk')
plt.legend()
plt.show()
```

Above is an example of a simple symmetric random walk

```python
##Compare multiple 1D Random Walks (Comparing 5 walks)
plt.figure(figsize=(10,6))
for i in range(5):
    steps=np.random.choice([-1,1], size=num_steps)
    positions=np.cumsum(steps)
    plt.plot(positions, label=f"Walk {i+1}", alpha=0.75)
```

```python
#2D Random Walk
num_step=1000
x_steps=np.random.choice([-1,1], size=num_step) # Steps in X-direction
y_steps=np.random.choice([-1,1], size=num_step) #Steps in y-direction
```

```python
x_positions=np.cumsum(x_steps) #Cumulative sum of x-steps
y_positions=np.cumsum(y_steps) #Cumulative sum of y-steps
```

```python
## 2D Random Walk
plt.figure(figsize=(10,6))
plt.plot(x_positions, y_positions, marker=".", linestyle="-", alpha=0.7)
plt.scatter(0,0, color="green", label="Start", zorder=3)
plt.scatter(x_positions[-1], y_positions[-1], color="blue", label="End", zorder = 3, s=100)
plt.legend()
plt.xlabel("X Position")
plt.ylabel("Y Position")
plt.title("2D Random Walk")
plt.grid()
plt.show()
```

## Probability

```python
# Higher Probability direction for positive direction
prob_right = 0.7 #70% chance of moving to the right direction.
steps = np.random.choice([-1,1], size=num_steps, p=[1-prob_right, prob_right])
positions = np.cumsum(steps)
```

```python
#plots
plt.figure(figsize=(10,5))
plt.plot(positions, label="Adjusted 1D Random Walk")
plt.axhline(y=0, color='black', linestyle='--', linewidth=0.8)
plt.xlabel("Steps")
plt.ylabel("Positions")
plt.title("Adjusted 1D Random Walk")
plt.legend()
plt.show()
```
