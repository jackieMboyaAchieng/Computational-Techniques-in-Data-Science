```python
# Fibonacci
def fibonacci_iterative(n):
    fibo = [0, 1]
    for i in range(2, n + 1):
        fibo.append(fibo[i - 1] + fibo[i - 2])
    return fibo[:n + 1]

print(fibonacci_iterative(12))
```

```python
# Fibonacci
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)
print("Fibonacci sequence up to 10 terms:")
for i in range(10):
    print(fibonacci(i), end=" ")
```

```python
#LIS Bottom up (DP)
def lis_dp(arr):
    n = len(arr)
    dp = [1] * n
    for i in range(1, n):
        for j in range(i):
            if arr[i] > arr[j]:
                dp[i] = max(dp[i], dp[j] + 1)
    return max(dp)
```

```python
num=[10, 9, 2, 5, 3, 7, 101, 18]
```

```python
toplist=lis_dp(num)
```

```python
#knapsack problem
def knapsack(W, wt, profit, n, memo=None):
    if memo is None:
        memo = {}
    if (n, W) in memo:
        return memo[(n, W)]
    if n == 0 or W == 0:
        return 0
    if wt[n-1] > W:
        result = knapsack(W, wt, profit, n-1, memo)
    else:
        result = max(profit[n-1] + knapsack(W-wt[n-1], wt, profit, n-1, memo),
                         knapsack(W, wt, profit, n-1, memo))
        memo [(n, W)] = result
        return result
```

```python
print(knapsack(50, [10, 20, 30], [60, 100, 120], 3))
```
